## Build Model and print the results non team input

```python
import pandas as pd
import numpy as np
import seaborn as sns
import sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("nfl_plays_offense_personnel.csv")
data
```

```python
data.dropna(subset=['down'], inplace=True)
data.dropna(subset=['play_type'], inplace=True)
data.dropna(subset=['score_differential'], inplace=True)
```

```python
def whatplay(yardline, down, mintues_left, qtr, yrdstogo, scoreDiff, posTimeouts, defTimeouts):
    data_predictors = data[['yardline_100', 
                        'down', 
                       'quarter_seconds_remaining', 
                        'qtr', 
                        'ydstogo',
                        'score_differential',
                        'posteam_timeouts_remaining',
                        'defteam_timeouts_remaining']]
    X = data_predictors
    y = data[['play_type']]

    rf_classifier = RandomForestClassifier(n_estimators=1000, random_state=42)
    rf_classifier.fit(X, y)
    X_predict = np.array([[yardline, down, mintues_left, qtr, yrdstogo, scoreDiff, posTimeouts, defTimeouts]])
    print(rf_classifier.predict(X_predict))
```

```python
whatplay(35, 2, 120, 2, 5, 3, 3, 3)
```

## Build Model and print the results Team input

```python
def whatplayTeam(team, yardline, down, mintues_left, qtr, yrdstogo, scoreDiff, posTimeouts, defTimeouts):
    data2 = data[data['posteam'] == team]
    data_predictors = data2[['yardline_100', 
                        'down', 
                       'quarter_seconds_remaining', 
                        'qtr', 
                        'ydstogo',
                        'score_differential',
                        'posteam_timeouts_remaining',
                        'defteam_timeouts_remaining']]
    X = data_predictors
    y = data2[['play_type']]

    rf_classifier = RandomForestClassifier(n_estimators=1000, random_state=42)
    rf_classifier.fit(X, y)
    X_predict = np.array([[yardline, down, mintues_left, qtr, yrdstogo, scoreDiff, posTimeouts, defTimeouts]])
    print(rf_classifier.predict(X_predict))
```

```python
whatplayTeam('DEN', 35, 2, 120, 2, 5, 3, 3, 3)
```
